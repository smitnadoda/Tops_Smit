```python
import numpy as np
import pandas as pd
his = pd.read_csv("historical_automobile_sales.csv")
his.head()
```

```python
his_apr = his["Month"] == "Apr"
apr = his[his_apr]
apr
```

```python
def his_month(Month):
    month_data = his["Month"] == Month
    return his[month_data]
his_month("Jan")
```

```python
his_month("Aug")
```

```python
def his_city_month(City,Month):
    city = his["City"] == City
    month = his["Month"] == Month
    city_month = city & month
    return his[city_month]
his_city_month('California','Apr').sort_values(by='Year',ascending=True).head(5)
his[['Month','City','Advertising_Expenditure']].plot(kind = 'bar')

```

```python
mon = his.groupby('Month')
mon['Competition'].sum().sort_values(ascending = False).plot(kind = 'bar')
```
